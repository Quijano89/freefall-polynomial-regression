## To Be or Not To Be
Write a program that prints "to be" if the sum of two variables is greater than 5 and "not to be" if less than 5.

```python
import time

def mabagal(text, delay=0.1):             # function
    for char in text:                     # To repeat character in text
        print(char, end='', flush=True)   # print character without a new line
        time.sleep(delay)                 # delay
    print()                               # new line

# Input Values
input_1 = input("Enter a number: ")
input_2 = input("Enter another number: ")

mabagal(f"You entered: {input_1} and {input_2}")

# Convert sila to float
input_1 = float(input_1)
input_2 = float(input_2)

# Operation
sum = input_1 + input_2

# Conditions and Print
if sum > 5:
  mabagal("to be")
elif sum < 5:
  mabagal("not to be")
else:
  exit()
```
